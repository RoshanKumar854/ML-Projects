```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x_1 = np.linspace(0, 100, 100)
x_2 = np.random.randint(0,100, 100)
x_3 = np.random.randint(20, 100, 100)
x_4 = np.random.uniform(0, 100, 100)
y = x_1+x_2+x_3+x_4
```

```python
fig = plt.figure(figsize = (12,8))
plt.subplot(2, 2, 1)
plt.scatter(x_1, y, color="red", marker="+")
plt.xlabel("nothing but just x-label!")

plt.subplot(2, 2, 2)
plt.scatter(x_2, y, color="purple", marker="_")

plt.subplot(2, 2, 3)
plt.scatter(x_3, y, color="pink", marker="*")

plt.subplot(2, 2, 4)
plt.scatter(x_4, y, color="green")

plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize = (12,8))
x = np.random.randint(0, 100, 100)
y = x**2
z = x**3
ax = plt.axes(projection='3d')
ax.scatter(x, y, z)

ax.set_title("3D graph")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

plt.show()
```

```python
fig = plt.figure(figsize = (12,8))
z = np.arange(0, 10*np.pi, np.pi/50)
x = np.sin(z)
y = np.cos(z)
ax = plt.axes(projection='3d')
ax.plot3D(x, y, z, color="red")

ax.set_title("3D graph")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

plt.show()
```

```python
fig = plt.figure(figsize = (12,8))
ax = plt.axes(projection='3d')
x = np.arange(-5, 5.1, 0.2)
y = np.arange(-5, 5.1, 0.2)

X, Y = np.meshgrid(x, y)
Z = np.sin(X)*np.cos(Y)

surf = ax.plot_surface(X, Y, Z, cmap = plt.cm.cividis)
fig.colorbar(surf, shrink=0.5, aspect=8)

plt.show()
```
