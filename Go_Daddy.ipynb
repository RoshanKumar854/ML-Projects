```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv("train.csv")
train_df = train_df.dropna()
```

```python
type(train_df["first_day_of_month"])
```

```python
# current date minus a year
from datetime import datetime
from dateutil.relativedelta import relativedelta
```

```python
train_df['Day Month'] = train_df["first_day_of_month"].str[4:]
train_df['Year'] = train_df["first_day_of_month"].str[:4]
```

```python
type(train_df['Year'])
```

```python
type(train_df['Year'])
```

```python
train_df['Year'] = pd.to_numeric(train_df['Year'])
train_df['Year'] = train_df['Year'] - 2
```

```python
train_df['Year']
```

```python
train_df['Month'] = train_df["first_day_of_month"].str[5:7]
```

```python
train_df['Month']
train_df['Month'] = pd.to_numeric(train_df['Month'])
```

```python
censor_df = pd.read_csv("census_starter.csv")
censor_df
```

```python
train_df = pd.merge(train_df, censor_df, on='cfips', how="inner")
```
