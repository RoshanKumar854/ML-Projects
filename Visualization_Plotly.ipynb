```python
import plotly.express as px
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/StudentsPerformance.csv")
df
```

```python
fig = px.scatter(df, x = "math score", y="reading score")
fig.show()
```

```python
fig = px.scatter(df, x = "math score", y="reading score", color="parental level of education")
fig.show()
```

```python
fig = px.box(df, y = "reading score")
fig.show()
```

```python
fig = px.box(df, y = "reading score", x = "gender", points = "all")
fig.show()
```

```python
fig = px.box(df, y = "reading score", x = "gender", points = "all", color = "parental level of education")
fig.show()
```

```python
fig = px.box(df, y = "reading score", points = "all", color = "parental level of education")
fig.show()
```

```python
fig = px.violin(df, y="reading score", box=True, points = "all", color = "parental level of education")
fig.show()
```

```python
fig = px.bar(df, x = "parental level of education", y = "math score")
fig.show()
```

```python
fig = px.bar(df, x = "parental level of education", y = "math score", color = "gender")
fig.show()
```

```python
fig = px.bar(df, x = "parental level of education", y = "math score", color = "gender", barmode = "group")
fig.show()
```

```python
fig = px.bar(df, x = "parental level of education", y = "math score", color = "gender", pattern_shape = "gender", pattern_shape_sequence = ["+", "x"])
fig.show()
```

```python
fig = px.scatter_matrix(df, dimensions=["writing score", "reading score", "math score"], color="gender")
fig.show()
```

```python
fig = px.pie(df, values="math score", names="parental level of education", title = "Math score based on parental education")
fig.show()
```

```python
fig = px.pie(df, values="math score", names="parental level of education", title = "Math score based on parental education", color_discrete_sequence = px.colors.sequential.RdBu)
fig.show()
```

```python
fig = px.pie(df, values="math score", names="parental level of education", title = "Math score based on parental education", color_discrete_map={"some college":"lightcyan",
                                 "associate's degree":"cyan",
                                 "high school":"royalblue",
                                 "some high school":"darkblue",
                                  "bachelor's degree":"red",
                                  "master's degree":"blue"})
fig.show()
```

```python
fig = px.histogram(df, x = "writing score")
fig.show()
```

```python
fig = px.histogram(df, x="parental level of education", y="math score", color="gender", marginal="rug", hover_data=df.columns)
fig.show()
```

```python
fig = px.strip(df, x="parental level of education", y = "writing score", orientation="v", color="gender", title = "i am not sure what title i should give to this chart")
fig.show()
```

```python
fig = px.ecdf(df, x = "math score", color="gender")
fig.show()
```

```python
fig = px.density_contour(df, x = "math score", y = "reading score", color="gender")
fig.show()
```

```python
fig = px.scatter_3d(df, x = "math score", y = "reading score", z= "writing score", color="parental level of education")
fig.show()
```

```python
fig = px.scatter_3d(df, x = "math score", y = "reading score", z = "writing score", color="gender", symbol ="parental level of education")
fig.show()
```
